# Mapping of University Names to Cantons

```python
# Do not forget to explain the format of the json
with open('json/ch-cantons.topojson.json') as json_data:
    cantons_data = json.load(json_data)

cantons_mappings = []
for i in range(0,len(cantons_data["objects"]["cantons"]["geometries"])):
    canton_id = cantons_data["objects"]["cantons"]["geometries"][i]["id"]
    canton_name = cantons_data["objects"]["cantons"]["geometries"][i]["properties"]["name"]
    cantons_mappings.append((canton_name,canton_id))
    
cantons_mappings_dict = dict(cantons_mappings)
cantons_mappings_dict.keys()
#cantons_mappings_dict['Genève']
```

```python
# DO NOT FORGET TO DELETE THIS KEY WHEN YOU PUSH
gmaps = googlemaps.Client(key='AIzaSyBwdIlLTYzo6Od_1qusJ2_hivcXFkqJzHs') 


# Geocoding an address
university_canton_result = []
for i in range(0,len(university_amount_df)):
    university_name = university_amount_df['University'][i]
    geocode_result = gmaps.geocode(university_name.split('-')[0]) 
    # We choose to do this splitting because it increases to the coverage from 18 to 29 
    if len(geocode_result)==0:
        university_canton_result.append((university_name,"Not Matching"))
    else:
        address_components = geocode_result[0]['address_components']
        flag = 0
        for j in range(0,len(address_components)):
            short_name = address_components[j]['short_name']
            if (short_name in cantons_mappings_dict.values()):
                flag = 1
                university_canton_result.append((university_name,short_name))
        if flag ==0:
            university_canton_result.append((university_name,"Not Matching"))
            
university_canton_result_dict = dict(university_canton_result)
```
